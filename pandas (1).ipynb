## Hierarchial indexing

```python
index = [('California', 2000),('California', 2001), ('California', 2010),
         ('New York', 2000), ('New York', 2010),
         ('Texas', 2000), ('Texas', 2010)]#list of tuples

population = [33871648,37253956,18976457,19378102,20851820,25145561,26448193]
```

```python
df = pd.Series(population,index=index)
df
```

```python
df[('California',2010)]
```

```python
df[('California',2010):('Texas',2010)]
```

```python
# converting a multiindex to sereis - done
# converting a multiindex to pandas dataframe

'''
1. Mutiindex setup
2. Reindexing
'''

#Multiindex setup
print(index)
index = pd.MultiIndex.from_tuples(index) #from list of tuples we are converting into MultiIndex

#Reindexing
print(df)
df = df.reindex(index)
print(df)
```

```python
index = [('California', 2000),('California', 2001), ('California', 2010),
         ('New York', 2000), ('New York', 2010),
         ('Texas', 2000), ('Texas', 2010)]#list of tuples

population = [33871648,37253956,18976457,19378102,20851820,25145561,26448193]

#we are converting this to series
df_series = pd.Series(population,index = index)
print(df_series)
index = pd.MultiIndex.from_tuples(index)
df_series = df_series.reindex(index)
print(df_series)
```

```python
df_series
```

```python
# I want to convert index 2 into the column
pop = df_series.unstack()
```
