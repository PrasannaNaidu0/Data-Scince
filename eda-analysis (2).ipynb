```python
'''
If there are lot of categories in the data, you need to consider the process of recategorization
'''

def mapping_education(x):
    
    if x in ['5th-6th', '10th', '1st-4th', 'Preschool', '12th','11th','9th','7th-8th']:
        return 'low_level_grade'
    elif x in ["HS-grad","Some-college",'Assoc-acdm', 'Assoc-voc']:
        return 'medium_level_grade'
    elif x in ['Bachelors','Prof-school','Masters','Doctorate']:
        return "high_level_grade"
```

```python
## declare a mapping funcation and perform apply
df['education_summary'] = df['education'].apply(mapping_education)
```

```python
sns.countplot(x='education_summary',data=df)
```

```python
df.groupby("education_summary")["salary"].value_counts()
```

```python
sns.countplot(x='education_summary',hue='salary',data=df)
```
